```python
# ---- A. Read the CSV File ----
csv_path = "F:\\01. The Data Psychology\\4. New_Machine Learning\\laptop price prediction\\20. Laptop_CSV_File.csv"
df = pd.read_csv(csv_path)
```

```python
# Optional: Rename 'Unnamed: 0' column to 'id' for clarity
df= df.iloc[:,1:]
```

```python
df.head(2)
```

```python
# ---- C. Create the Table ----
create_table_query = """
CREATE TABLE IF NOT EXISTS laptops (
    Company VARCHAR(255),
    TypeName VARCHAR(255),
    Inches FLOAT,
    ScreenResolution VARCHAR(255),
    Cpu VARCHAR(255),
    Ram VARCHAR(20),
    Memory VARCHAR(255),
    Gpu VARCHAR(255),
    OpSys VARCHAR(255),
    Weight VARCHAR(20),
    Price FLOAT
)
"""
```

```python
mycursor.execute(create_table_query)
```

```python
insert_query = """
INSERT INTO laptops (Company, TypeName, Inches, ScreenResolution, Cpu, Ram, Memory, Gpu, OpSys, Weight, Price)
VALUES ( %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
ON DUPLICATE KEY UPDATE 
    Company=VALUES(Company), TypeName=VALUES(TypeName), Inches=VALUES(Inches),
    ScreenResolution=VALUES(ScreenResolution), Cpu=VALUES(Cpu), Ram=VALUES(Ram),
    Memory=VALUES(Memory), Gpu=VALUES(Gpu), OpSys=VALUES(OpSys), Weight=VALUES(Weight), Price=VALUES(Price)
"""
```

```python
# Prepare data as tuples
data_tuples = [tuple(row) for row in df.values]
```

```python
data_tuples
```

```python
# Insert all rows efficiently
mycursor.executemany(insert_query, data_tuples)
connection.commit()
```
